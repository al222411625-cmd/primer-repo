```python
# Programa de sueldos de empleados

while True:
    print("MENÚ DE SUELDOS")
    print("1. RECTOR")
    print("2. DIRECTOR")
    print("3. PROFESOR DE TIEMPO COMPLETO")
    print("4. PROFESOR DE ASIGNATURA")
    print("5. Salir")

    opcion = input("Elige una opción (1-5): ")

    if opcion == "5":
        print("Programa terminado.")
        break


    if opcion == "1":
        pago_hora = 800
        nombre = "Rector"
    elif opcion == "2":
        pago_hora = 400
        nombre = "Director"
    elif opcion == "3":
        pago_hora = 120
        nombre = "Profesor de Tiempo Completo"
    elif opcion == "4":
        pago_hora = 110
        nombre = "Profesor de Asignatura"
    else:
        print("Opción no válida, intenta de nuevo.")
        continue

    # Pedir horas
    horas_normales = int(input("Horas trabajadas normales: "))
    horas_extra_doble = int(input("Horas extra dobles: "))
    horas_extra_triple = int(input("Horas extra triples: "))

    # Calcular sueldo
    sueldo = (horas_normales * pago_hora) + \
           (horas_extra_doble * pago_hora * 2) + \
           (horas_extra_triple * pago_hora * 3)

    print(f"El sueldo del {nombre} es: ${sueldo}")
```

```python
# Diccionario con 10 países de Asia y sus capitales
asia = {
    "China": "Pekín",
    "Japón": "Tokio",
    "India": "Nueva Delhi",
    "Corea del Sur": "Seúl",
    "Tailandia": "Bangkok",
    "Vietnam": "Hanói",
    "Indonesia": "Yakarta",
    "Filipinas": "Manila",
    "Arabia Saudita": "Riad",
    "Israel": "Jerusalén"
}

while True:
    print("MENU DE SALIDA:")
    print("1.- Buscar por capital")
    print("2.- Buscar por país")
    print("3.- Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        cap = input("Escribe la capital: ")
        encontrado = False
        for pais, capital in asia.items():
            if capital.lower() == cap.lower():
                print("La capital", capital, "es de", pais)
                encontrado = True
                break
        if not encontrado:
            print("No encontrada.")
    elif opcion == "2":
        pais = input("Escribe el país: ")
        if pais in asia:
            print("La capital de", pais, "es", asia[pais])
        else:
            print("No encontrado.")
    elif opcion == "3":
        print("Adiós")
        break
    else:
        print("Opción no válida")
```
